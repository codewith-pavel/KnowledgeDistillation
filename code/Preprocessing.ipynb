```python
import tensorflow as tf
import os
import numpy as np
from PIL import Image
from keras.preprocessing.image import ImageDataGenerator
import shutil
from keras.utils import to_categorical
import cv2
import matplotlib.pyplot as plt
```

```python
# Define the data augmentation transformations
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    zoom_range=0.1,
    shear_range=0.1,
    fill_mode='nearest'
)
```

```python
directory = 'NSCLC/Normal'  # replace with the path to your data directory
save_dir = 'NSCLC/Normal'  # replace with the path to the directory where you want to save augmented images
max_images = 5000  # set the maximum number of augmented images to generate
i = 0

for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.png'):  # or .jpg, .jpeg, etc.
            img_path = os.path.join(root, file)
            img = Image.open(img_path)
            img_array = np.array(img)

            for batch in datagen.flow(img_array.reshape(1, *img_array.shape), batch_size=1, save_to_dir=save_dir, save_prefix='Adenocarcinoma_', save_format='png'):
                i += 1
                if i >= max_images:
                    break  # stop the loop if the maximum number of images has been generated
            if i >= max_images:
                break  # stop the outer loop as well
    if i >= max_images:
        break
```

```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

src_dir = "NSCLC_Final"
dest_dir = "NSCLC_Radiomics_50k_TTV"
os.makedirs(dest_dir, exist_ok=True)

classes = ['Adenocarcinoma', 'Large_cell_carcinoma', 'Normal', 'Not_otherwise_specified', 'Squamous_cell_carcinoma']
ratios = [0.4, 0.3, 0.3]

# Create train, test, and validation folders in the destination directory
for split in ["train", "test", "validation"]:
    os.makedirs(os.path.join(dest_dir, split), exist_ok=True)
    for c in classes:
        os.makedirs(os.path.join(dest_dir, split, c), exist_ok=True)

for c in classes:
    image_files = [f for f in os.listdir(os.path.join(src_dir, c)) if f.endswith('.png')]

    # Apply stratified sampling to split the data
    train_files, temp_files = train_test_split(image_files, train_size=ratios[0], stratify=[c] * len(image_files))
    test_files, val_files = train_test_split(temp_files, test_size=ratios[1] / (ratios[1] + ratios[2]), stratify=[c] * len(temp_files))

    for split, files in [("train", train_files), ("test", test_files), ("validation", val_files)]:
        for f in files:
            src_file = os.path.join(src_dir, c, f)
            dest_file = os.path.join(dest_dir, split, c, f)
            shutil.copy(src_file, dest_file)

print("Images organized successfully!")
```

```python
# Loop through the train, test, and validation sets
for split in ["train", "test", "validation"]:
    print(f"Images in {split} set:")
    for c in classes:
        class_dir = os.path.join(dest_dir, split, c)
        image_files = os.listdir(class_dir)
        for f in image_files:
            print(f"Class: {c}, File: {f}")
    print()
```

```python
dest_dir = "NSCLC_Radiomics_50k_TTV"
# Calculate the percentage distribution of each class in train, test, and validation sets
for split in ["train", "test", "validation"]:
    print(f"Class distribution in {split} set:")
    total_samples = 0
    class_samples = {c: 0 for c in classes}

    for c in classes:
        class_dir = os.path.join(dest_dir, split, c)
        num_samples = len(os.listdir(class_dir))
        class_samples[c] = num_samples
        total_samples += num_samples

    for c in classes:
        percentage = (class_samples[c] / total_samples) * 100
        print(f"{c}: {percentage:.2f}%")
    print()
```
